### 捐款企業、現金與是否當選

```python
df[["Name","Donor","Amount","Elected"]].groupby("Name").agg(
                                                    {"Donor":"count",
                                                     "Amount":"sum",
                                                    "Elected": lambda x: 1 if x.sum() > 1 else 0
                                                    }
                                                    ).sort_values(["Donor","Elected","Amount"],ascending = [False,True,False])
```
