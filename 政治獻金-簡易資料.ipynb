```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv(r".\try1.csv")
df1 = pd.read_csv(r".\user.csv")
```

```python
display(df.head(5),df1.head(5))
```

```python
pd.set_option('display.max_rows', 1000)
```

### 捐款企業、現金與是否當選

```python
data = df[["Name","Donor","Amount","Elected"]].groupby("Name").agg(
                                                    {"Donor":"count",
                                                     "Amount":"sum",
                                                    "Elected": lambda x: 1 if x.sum() > 1 else 0
                                                    }
                                                    ).sort_values(["Donor","Elected","Amount"],ascending = [False,True,False])
```

```python
display(data)
```

### Visualization

以金額的角度來看，會不會當選的明顯分界點在一千萬

```python
plt.figure(figsize=(20,10))
sns.stripplot(data=data, x="Elected" , y="Amount",jitter= .4, size = 4)
plt.show()
```

以捐款的企業數來看，會不會當選的明顯分界在75個

```python
plt.figure(figsize=(20,10))
sns.stripplot(data=data, x="Elected" , y="Donor",jitter= .4, size = 4)
plt.show()
```
